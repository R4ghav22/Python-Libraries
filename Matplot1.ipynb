```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(3,2))
graph=plt.plot([1,2,3] , [10,20,30])
```

```python
plt.figure(figsize=(5,3))
graph=plt.plot([1,2,3,4,5] , [10,20,65,93,72] , color="red" , linestyle="solid" ,marker="h") 
plt.title("Line Plot")
plt.xlabel("Players")
plt.ylabel("Scores")

                
```

```python
plt.figure(figsize=(5,3))
plt.scatter([1,2,3,4,5] , [10,20,65,93,72] , color="black"  ,marker="h")
plt.title("Line Plot")
plt.xlabel("Players")
plt.ylabel("Scores")

```

```python
plt.figure(figsize=(5,3))
plt.bar([1,2,3,4,5] , [10,20,65,93,72] , color="black")
plt.title("Line Plot")
plt.xlabel("Players")
plt.ylabel("Scores")

```

```python
plt.figure(figsize=(5,3))
plt.hist([10,30,65,93,72,35, 45, 21, 27, 58,75,76,77, 81], color='yellow')
plt.title('Histogram')
plt.xlabel('scores')
plt.ylabel('players')
```

```python
plt.figure(figsize=(6,4))
plt.pie([10,31,76,89,43] , labels= ['A' ,'B' , 'C' ,'D' , 'E' ] , autopct="%1.2f%%") 
plt.savefig('pie-chart')
```
