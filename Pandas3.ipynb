```python
import pandas as pd


```

```python
data = pd.DataFrame({
    "Category": ["Furniture", "Technology", "Furniture", "Technology", "Office"],
    "Sales": [1000, 2000, 1500, 3000, 800]
})
data
```

```python
data['Category']
```

```python
data.groupby('Category')['Sales'].sum()
```

```python
data.groupby('Category')['Sales'].mean()
```

```python
data.groupby('Category')['Sales'].max()
```

```python
data.groupby('Category')['Sales'].count()
```
