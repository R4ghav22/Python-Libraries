```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({'name':['Arjun', 'Arun', 'Keshav'], 'age':[23,31,25], 'country':['India', 'Canada', 'France']})
df
```

```python
sns.scatteredplot(x='age', y='country', data=df)
plt.show()
```

```python
names = sns.get_dataset_names()
print(names)
```

```python
tips= sns.load_dataset("tips")
tips
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="total_bill", y="tip", hue="sex", size="size", sizes=(30,50), marker='*', data=tips)
plt.xlabel("total Bill")
plt.ylabel("Tip")
plt.title("Relationship b/w total bill and tip ")
plt.show()
```
