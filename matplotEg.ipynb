```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(3,2))
line=plt.plot([1,2,3,4], [10,20,30,40])
# plt.getp(line)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.plot([1,2,3,4,5],[10,30,65,93,72],color='g', linestyle='solid',marker='*', linewidth='1')
plt.title('Line Plot')
plt.xlabel('player')
plt.ylabel('scores')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.scatter([1,2,3,4,5],[10,30,65,93,72],color='green',linewidths=0.5)
plt.title('Scatter Plot')
plt.xlabel('player')
plt.ylabel('scores')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.bar([1,2,3,4,5],[10,30,65,93,72],color='orange')
plt.title('Bar Chart')
plt.xlabel('player')
plt.ylabel('scores')
plt.savefig('barchart.png')
plt.show()
```

```python
# visualizing the distribution of a dataset
plt.figure(figsize=(5,3))
plt.hist([10,30,65,93,72,35, 45, 21, 27, 58,75,76,77, 81], color='yellow')
plt.title('Hitogram')
# plt.xlabel('player')
# plt.ylabel('scores')
plt.show()
```

```python
plt.figure(figsize=(9,5))
size = [10,30,65,93,72]
labels = ['A','B', 'C', 'D', 'E']
plt.pie(size,labels=labels,a2utopct='%1.2f%%')
plt.savefig('pie-chart')
plt.title('Pie Chart')
plt.show()
```
